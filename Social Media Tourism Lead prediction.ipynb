# Outlier Treatment 

```python
# taking only those columns which do not have categorical values
```

```python
list = ['Yearly_avg_view_on_travel_page','total_likes_on_outstation_checkin_given',
       'yearly_avg_Outstation_checkins', 'Yearly_avg_comment_on_travel_page',
        'total_likes_on_outofstation_checkin_received','montly_avg_comment_on_company_page',
        'Daily_Avg_mins_spend_on_traveling_page']
```

```python
plt.figure(figsize=(15,12))
df[list].boxplot(vert=0)
plt.title("With Outliers")
plt.show()
```

```python
sorted(list)
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in df[list].columns:
    lr,ur=remove_outlier(df[column])
    df[column]=np.where(df[column]>ur,ur,df[column])
    df[column]=np.where(df[column]<lr,lr,df[column]) #replacing values
```

```python
plt.figure(figsize=(15,12))
df[list].boxplot(vert=0)
plt.title('After Outlier Removal',fontsize=16)
plt.show()
```
