```python
model = Sequential()
model.add(Dense(units=26,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=13,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=6,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(optimizer='adam',loss = 'binary_crossentropy')
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
fpr,tpr, thresholds = roc_curve(y_test, proba_log[:,1] )
```

```python
def plot_roc_curve(fpr,tpr):
    plt.plot(fpr,tpr ,color = 'orange', label = 'ROC')
    plt.plot([0,1],[0,1],color = 'blue')
    plt.xlabel('fpr')
    plt.ylabel('tpr')
    plt.legend()
```

```python
plot_roc_curve(fpr,tpr)
```

```python
print(roc_auc_score(y_test,proba_log[:,1]))
```

```python
from sklearn.metrics import accuracy_score
accuracy_is = []
for thres in thresholds:
    y_pred = np.where(proba_log[:,1]>thres,1,0)
    accuracy_is.append(accuracy_score(y_test,y_pred, normalize= True))
    
accuracy_is = pd.concat([pd.Series(thresholds),pd.Series(accuracy_is)],axis=1)
accuracy_is.columns =['thresholds','accuracy']
accuracy_is.sort_values(by='accuracy',ascending=False,inplace=True)
accuracy_is.head()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
```

```python
roc_predictions = [1 if i >= 0.623919 else 0 for i in proba_log[:,1] ]
```

```python
print("Accuracy Score Before and After Thresholding: {}, {}".format(accuracy_score(y_test, pred_log), accuracy_score(y_test, roc_predictions)))
print("Precision Score Before and After Thresholding: {}, {}".format(precision_score(y_test, pred_log), precision_score(y_test, roc_predictions)))
print("Recall Score Before and After Thresholding: {}, {}".format(recall_score(y_test, pred_log), recall_score(y_test, roc_predictions)))
print("F1 Score Before and After Thresholding: {}, {}".format(f1_score(y_test, pred_log), f1_score(y_test, roc_predictions)))
```

```python
print(classification_report(y_test,roc_predictions))
print(confusion_matrix(y_test,roc_predictions))
```

# Bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
X = df.drop('Approved',axis=1)

y= df['Approved']



from imblearn.over_sampling import SMOTE

sm = SMOTE()

X_sm, y_sm = sm.fit_resample(X, y)

print(y.value_counts())
print(y_sm.value_counts())

from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.3, random_state=101)



# from sklearn.model_selection import train_test_split 

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
base_classifier = DecisionTreeClassifier()
```

```python
bagging_classifier = BaggingClassifier(base_estimator=base_classifier, n_estimators=50, random_state=101)
```

```python
bagging_classifier.fit(X_train,y_train)
```

```python
pred_bagging = bagging_classifier.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred_bagging))
print(classification_report(y_test,pred_bagging))
print(accuracy_score(y_test, pred_bagging))
```
