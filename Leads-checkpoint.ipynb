```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
probs = model.predict_proba(X_test)
```

```python
probs
```

```python
fpr, tpr, thresholds = roc_curve(y_test,probs[:,1])
```

```python
def plot_roc_curve(fpr,tpr):

    plt.plot(fpr,tpr, color = 'orange', label = 'ROC')
    plt.plot([0,1],[0,1], color = 'blue')
    plt.xlabel("false positive rate")
    plt.ylabel("true positive rate")
    plt.legend()
```

```python
plot_roc_curve(fpr,tpr)
```

```python
print(roc_auc_score(y_test,probs[:,1]))
```

```python
from sklearn.metrics import accuracy_score
accuracy_is = []
for thres in thresholds:
    y_pred = np.where(probs[:,1]>thres,1,0)
    accuracy_is.append(accuracy_score(y_test,y_pred, normalize= True))
    
accuracy_is = pd.concat([pd.Series(thresholds),pd.Series(accuracy_is)],axis=1)
accuracy_is.columns =['thresholds','accuracy']
accuracy_is.sort_values(by='accuracy',ascending=False,inplace=True)
accuracy_is.head()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
```

```python
roc_predictions = [1 if i >= 0.424586 else 0 for i in probs[:,1] ]
```

```python
print("Accuracy Score Before and After Thresholding: {}, {}".format(accuracy_score(y_test, prediction), accuracy_score(y_test, roc_predictions)))
print("Precision Score Before and After Thresholding: {}, {}".format(precision_score(y_test, prediction), precision_score(y_test, roc_predictions)))
print("Recall Score Before and After Thresholding: {}, {}".format(recall_score(y_test, prediction), recall_score(y_test, roc_predictions)))
print("F1 Score Before and After Thresholding: {}, {}".format(f1_score(y_test, prediction), f1_score(y_test, roc_predictions)))
```

```python
print('Before thresholding')
print(confusion_matrix(y_test,prediction))
print(classification_report(y_test,prediction))

print('After thresholding')
print(confusion_matrix(y_test,roc_predictions))
print(classification_report(y_test,roc_predictions))
```
